The following statistical model describes and predicts CO2 data measured in Mauna Loa between 1958 - 2018. It models the overall trend, seasonal variation and noise. Using priors for the parameters and likelihood of the data, posterior distributions are calculated and used to predict atmospheric CO2 levels until 2058. The overall trend is quadratic, with a tilted sine function to model variational change. The code is in Python and PyStan.

### The Model

<center>
**Model Equation** 

<center>
p(x<sub>t</sub>|$\theta) = N(c_0 + c_1t + c_2t^2 + c_3 \frac{1}{c_4} \arctan(\frac{c_4\sin(\frac{2\pi t}{365.25})+c_5}{1-c_4 \cos(\frac{2 \pi t}{365.25})+c_5}), \sigma)$
    
</center>

<center>
**Variables for time trend**

variable    |    limits    |    prior    |    meaning
--|--|--|--
c<sub>0</sub>|> 0|Normal(310,20)|The constant determines the intercept, and should be somewhere <br/> close to 300, which was approximately the level in 1958.
c<sub>1</sub>|> 0|Normal(1,2)|The linear term is difficult to estimate as it might be offset by the quadratic <br/> term, but it is positive and should not be larger than 4, as that would be too steep.
c<sub>2</sub>|> 0|Normal(1,2)|Similarly, the quadratic term is difficult to estimate and unknown to me. Therefore, the prior is large, centered on 1. 

<br/>

<center>
**Variables for seasonal changes**
<center>
Experimented with on Desmos (https://www.desmos.com/calculator) using the equation 
<center> 
$a\frac{1}{t}\arctan\left(\frac{t\sin\left(x-d\right)}{1-t\ \cos\left(x-d\right)}\right)$

variable    |    limits    |    prior    |    meaning
--|--|--|--
c<sub>3</sub>|> 0|Normal(3,2)|The amplitude changes the y-range of the function. It should be positive since it <br/> otherwise flips, and below 5, as it otherwise would be too steep.
c<sub>4</sub>| -1 < c<sub>4</sub> > 0|Cauchy(-0.5,1)| If this parameter is outside (-1,1) it is not continuous, and flips to be left tilted between 0 and 1. <br/> Since I don't know more about it, I set a broad prior centered on -0.5.
c<sub>5</sub>|<0|Normal(0,1)| Prior knowledge about the phase is difficult to incorporate, thus a rather broad prior is chosen for values around 1.
sigma $\sigma$ |> 0|Normal(1.5,1)|The standard deviation models the noise we can have in our data. An estimate for a day could be for example 365.5 ppm, but probably can vary on average by 1-2 ppms for a given observation. I don't expect the variation to be too large as the year by year change itself is not too large.


Note: I arrived at this equation to model seasonal change via a discussion on Stack Exchange here:
https://math.stackexchange.com/questions/2430564/equation-of-a-tilted-sine
<center>
$a\frac{1}{t}\arctan\left(\frac{t\sin\left(x-d\right)}{1-t\ \cos\left(x-d\right)}\right)$
</center>

### Time Trend Model

```python
# FINAL MODEL (from before)

stan_code = """
data {
    int<lower=0> n;        // The number of data
    real level[n];          // The levels / data
    real time[n];          // time days
}

parameters {

    real<lower=0> c_0;
    real<lower=0> c_1;
    real<lower=0> c_2;
    real<lower=0> sigma;
}

model {
    
    c_0 ~ normal(310, 20);
    c_1 ~ normal(1, 2);
    c_2 ~ normal(1, 2);
    sigma ~ normal(1.5, 1);
    
    for(i in 1:n) {
        level[i] ~ normal(
            c_0 + c_1 * time[i] + c_2 * time[i]^2
            , sigma);
    }
}

"""

stan_model = pystan.StanModel(model_code=stan_code)
```

```python
# Specify data 
model_n = 200
level_data = original_data['level'][:model_n]
date_data = original_data['delta_days'][:model_n]
n = len(level_data)

# Specify data for Stan
stan_data = {
    'n': n,
    'level': level_data,
    'time': date_data} 

# Print results summary from model
results = stan_model.sampling(data=stan_data)
samples = results.extract()
parameters = ['c_0', 'c_1', 'c_2','sigma']
print(results.stansummary(pars=parameters))
```

```python
time = original_data['delta_days'][:model_n]

c_0 = np.mean(samples['c_0'])
c_1 = np.mean(samples['c_1'])
c_2 = np.mean(samples['c_2'])
sigma = np.mean(samples['sigma'])

modeled_means = []

# plot modeled data
for i in range(model_n):
    modeled_means.append(c_0 + c_1 * time[i] + c_2 * time[i]**2)
plt.plot(modeled_means, label='model')

# plot actual data
plt.plot(original_data['level'][:(model_n)], label='data')

plt.legend()
plt.title('Quadratic Trend')
plt.plot()
```

```python
actual_data = original_data['level'][:200]
modeled_data = modeled_means[:200]
residuals = actual_data-modeled_data

plt.figure(figsize=(7,5))
plt.plot(residuals)
plt.title('Residuals')
plt.show()
```

### Final Model

```python
stan_code = """
data {
    int<lower=0> n;        // The number of data
    real level[n];          // The levels / data
    real time[n];          // time days
    int<lower=0> n_future; // The number of future values to predict
}

parameters {

    real<lower=0> c_0;
    real<lower=0> c_1;
    real<lower=0> c_2;
    real<lower=0> c_3;
    real<lower=-0.9, upper=-0.5> c_4;
    real<lower=0> c_5;
    real<lower=0> sigma;
}

model {
    
    c_0 ~ normal(310, 20);
    c_1 ~ normal(1, 2);
    c_2 ~ normal(1, 2);
    c_3 ~ normal(3, 2);
    c_4 ~ cauchy(-0.5, 1);
    c_5 ~ normal(1,1);
    sigma ~ normal(1.5, 1);
    
    for(i in 1:n) {
        level[i] ~ normal(
            c_0 + c_1 * time[i] + c_2 * time[i]^2
            
            + c_3 * 1/c_4 * atan((c_4 * sin(2 * pi() * time[i] / 365.25 + c_5))/
                                    (1 - c_4 * cos(2 * pi() * time[i] / 365.25 + c_5)))
            
            , sigma);
    }
}

// Generate the predicted function values for the next n_future steps.
generated quantities {
    
    real x_future[n_future];
    
    for(t in 1:n_future) {
        x_future[t] = normal_rng(
            c_0 + c_1 * (7*t + time[n]) + c_2 * (7*t + time[n])^2
            + c_3 * 1/c_4 * atan((c_4 * sin(2 * pi() * (7*t + time[n]) / 365.25 + c_5))/
                                    (1 - c_4 * cos(2 * pi() * (7*t + time[n]) / 365.25 + c_5)))
            , sigma);
    }
}

"""

stan_model = pystan.StanModel(model_code=stan_code)
```

### Test model on validation set

```python
# Specify data 
model_n = 1000
predict_n = 500

level_data = original_data['level'][:model_n]
date_data = original_data['delta_days'][:model_n]
n = len(level_data)
n_future = predict_n

# Specify data for Stan
stan_data = {
    'n': n,
    'level': level_data,
    'time': date_data,
    'n_future': n_future} 

# Print results summary from model
results = stan_model.sampling(data=stan_data)
samples = results.extract()
parameters = ['c_0', 'c_1', 'c_2', 'c_3', 'c_4', 'c_5','sigma']
print(results.stansummary(pars=parameters))
```

```python
time = original_data['delta_days'][:model_n]

c_0 = np.mean(samples['c_0'])
c_1 = np.mean(samples['c_1'])
c_2 = np.mean(samples['c_2'])
c_3 = np.mean(samples['c_3'])
c_4 = np.mean(samples['c_4'])
c_5 = np.mean(samples['c_5'])
sigma = np.mean(samples['sigma'])

modeled_means = []

plt.figure(figsize=(15,15))


# plot actual data
plt.plot(original_data['level'][:(model_n+predict_n)], label='data', color='r', alpha=0.5, linewidth=0.8)

# plot modeled data
for i in range(model_n):
    modeled_means.append(c_0 + c_1 * time[i] + c_2 * time[i]**2
                        + c_3 * 1/c_4 * math.atan((c_4 * math.sin(2 * math.pi * time[i] / 365.25 + c_5)) /
                                                  (1 - c_4 * math.cos(2 * math.pi * time[i] / 365.25 + c_5))))
plt.plot(modeled_means, label='model', color='g', alpha=0.5, linewidth=0.8)

# plot prediction
x = [i for i in range(model_n,model_n+predict_n)]
mean_predictions = samples['x_future'].mean(axis=0)
plt.plot(x, mean_predictions, label='prediction', color='b',alpha=0.5, linewidth=0.8)
    

plt.legend()
plt.title('Validate prediction on 500 datapoints')
plt.plot()
```

# Model Assessment

### Pairplots
In the pairplot graph below we can see that all parameters are normally distributed, with the exception of c_4, which was restricted to be between -0.5 and -1 and thus portrays a truncated distribution. Furthermore, the constant c_0 is negatively correlated with the linear term c_1 and positively correlated with the quadratic term c_2. The negative relationship between c_0 and c_1 is reasonable, as a higher intercept models the data better with a lower linear incline. I would not have expected the constant to be positively correlated with the quadratic term, but apparently, a higher intercept models the data better with a higher quadratic term. 

```python
# Make pair plot of the posteriors over all parameters of the model.
df = pd.DataFrame(
    data=np.transpose([samples[param] for param in parameters]),
    columns=parameters)
sns.pairplot(df, size=2.5, plot_kws={'marker': '.', 'alpha': 0.25})
plt.show()
```

### Autocorrelation Graph

The autocorrelation graph shows that there is almost no autocorrelation among the samples, indicating that the chains mixed well and we have a rather high level of independent samples. This is also visible in the output above, where the effective sample size is above 1000 in the worst case (c_4), and well above 2000 in most cases. 

```python
def plot_acf(x):
    '''
    Plot the autocorrelation function for a series x. This corresponds to the
    acf() function in R. The series x is detrended by subtracting the mean of
    the series before computing the autocorrelation.
    '''
    from scipy import signal
    plt.acorr(
        x, maxlags=20, detrend=lambda x: signal.detrend(x, type='constant'))

for param in parameters:
    plt.figure(figsize=(12, 4))
    plot_acf(samples[param])
    plt.title(f'Autocorrelation of {param} samples')

plt.show()
```

### Model with simple priors

```python
stan_code = """
data {
    int<lower=0> n;        // The number of data
    real level[n];          // The levels / data
    real time[n];          // time days
    int<lower=0> n_future; // The number of future values to predict
}

parameters {

    real<lower=0> c_0;
    real<lower=0> c_1;
    real<lower=0> c_2;
    real<lower=0> c_3;
    real<lower=-0.9, upper=-0.5> c_4; 
    real<lower=0> c_5;
    real<lower=0> sigma;
}

model {
    
    c_0 ~ normal(0, 5);
    c_1 ~ normal(0, 1);
    c_2 ~ normal(0, 1);
    c_3 ~ normal(0, 1);
    c_4 ~ normal(-0.7, 0.5);
    c_5 ~ normal(0, 1);
    sigma ~ normal(0, 1);
    
    for(i in 1:n) {
        level[i] ~ normal(
            c_0 + c_1 * time[i] + c_2 * time[i]^2
            
            + c_3 * 1/c_4 * atan((c_4 * sin(2 * pi() * time[i] / 365.25 + c_5))/
                                    (1 - c_4 * cos(2 * pi() * time[i] / 365.25 + c_5)))
            
            , sigma);
    }
}

// Generate the predicted function values for the next n_future steps.
generated quantities {
    
    real x_future[n_future];
    
    for(t in 1:n_future) {
        x_future[t] = normal_rng(
            c_0 + c_1 * (7*t + time[n]) + c_2 * (7*t + time[n])^2
            + c_3 * 1/c_4 * atan((c_4 * sin(2 * pi() * (7*t + time[n]) / 365.25 + c_5))/
                                    (1 - c_4 * cos(2 * pi() * (7*t + time[n]) / 365.25 + c_5)))
            , sigma);
    }
}

"""

stan_model = pystan.StanModel(model_code=stan_code)
```

```python
# Specify data 
level_data = original_data['level'][:model_n]
date_data = original_data['delta_days'][:model_n]
n = len(level_data)
n_future = predict_n

# Specify data for Stan
stan_data = {
    'n': n,
    'level': level_data,
    'time': date_data,
    'n_future': n_future} 

# Print results summary from model
results = stan_model.sampling(data=stan_data)
samples = results.extract()
parameters = ['c_0', 'c_1', 'c_2', 'c_3', 'c_4', 'c_5','sigma']
print(results.stansummary(pars=parameters))
```
